```python
import geopandas as gpd
from pyproj import Transformer
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

# first transform everything to standard coordinates

```python
# INDOOR
# Load the CSV file into a GeoDataFrame
gdf = gpd.read_file('../data/Galv_Bins_July_Version/indoor.csv')

# Define the source and target coordinate systems
source_crs = 'EPSG:3857'  # Web Mercator (WGS_1984_Web_Mercator_Auxiliary_Sphere)
target_crs = 'EPSG:4326'  # WGS84 (decimal degrees)

# Create a transformer to handle the coordinate conversion
transformer = Transformer.from_crs(source_crs, target_crs, always_xy=True)

# Apply the coordinate transformation to all points
gdf['longitude'], gdf['latitude'] = transformer.transform(gdf['x'].values, gdf['y'].values)

# Now, your DataFrame contains the points in latitude-longitude format (WGS84 decimal degrees)
gdf.head()
```

```python
# OUTDOOR
# Load the CSV file into a GeoDataFrame
gdf2 = gpd.read_file('../data/Galv_Bins_July_Version/outdoor.csv')

# Define the source and target coordinate systems
source_crs = 'EPSG:3857'  # Web Mercator (WGS_1984_Web_Mercator_Auxiliary_Sphere)
target_crs = 'EPSG:4326'  # WGS84 (decimal degrees)

# Create a transformer to handle the coordinate conversion
transformer = Transformer.from_crs(source_crs, target_crs, always_xy=True)

# Apply the coordinate transformation to all points
gdf2['longitude'], gdf2['latitude'] = transformer.transform(gdf2['x'].values, gdf2['y'].values)

# Now, your DataFrame contains the points in latitude-longitude format (WGS84 decimal degrees)
gdf2.head()
```

```python
# COMBINE INDOOR AND OUTDOOR INTO A SINGLE GEODATAFRAME WITH "LONGITUDE" AND "LATITUDE" COLUMNS ONLY
# CLEAN UP AND GIVE EACH ROW A UNIQUE OBJECTID

data = gdf.append(gdf2, ignore_index=True)
data = data[['OBJECTID', 'longitude', 'latitude']]
data['OBJECTID'][:30] = [i for i in range(1001, 1031)]
data['OBJECTID'][30:] = [i for i in range(2001, 2129)]
data.head()
```
